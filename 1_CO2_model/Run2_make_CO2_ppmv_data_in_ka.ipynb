```python
# read in CO2_sspxxx_ppmv.txt and get year 0, 1000, 2000, ... values only
# then save as CO2_data_sspxxx_0-1ma.res

import numpy as np
scenarios = ['SSP5-85', 'SSP3-70', 'SSP2-45','SSP1-26','SSP1-19','10000pgc']

for scenario in scenarios:
    CO2_timeseries = np.loadtxt('SSP/concentration_ppmv/CO2_' + scenario + '_ppmv.txt', dtype=float, comments="#", skiprows=0)
    time_full = CO2_timeseries[:,0]
    co2_full = CO2_timeseries[:,1]
    time_0_1ma = np.arange(0, 1001, 1)  # time from 0 to 1000 ka at 1 ka interval
    timeinfull = time_0_1ma * 1000  # convert to years
    mask = np.isin(time_full, timeinfull)
    co2_0_1ma = co2_full[mask]
    co2_0_1ma_timeseries = np.column_stack((time_0_1ma, co2_0_1ma))
    co2_0_1ma_text = 'Time_kyr_AP CO2_ppmv'
    # np.savetxt('../Conceptual-GSL/PosivaSKB-master/ForcingData/CO2_data_' + scenario + '_0-1ma.res', co2_0_1ma_timeseries, fmt = '%.6g', delimiter = ' ', newline = '\n', header = co2_0_1ma_text)
    path="SSP/concentration_ppmv/"
    np.savetxt(path+'CO2_data_' + scenario + '_0-1ma.res', co2_0_1ma_timeseries, fmt = ['%d','%.2f'], delimiter = ' ', newline = '\n', header = co2_0_1ma_text)
    print('Saved CO2_data_' + scenario + '_0-1ma.res')
```
