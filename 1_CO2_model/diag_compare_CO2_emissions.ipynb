This script is used to compare CO2 in ppmv in the same timeline for different scenarios

```python
sce="SSP1-26"
base_file=f"/Users/bo20541/Library/CloudStorage/OneDrive-UniversityofBristol/TONIC-Oligocene/Emulator_Xin/2025_Bristol_5D_v002_Xin/CO2_calculation/SSP/CO2_{sce}_ppmv.txt"
new_file=f"SSP/CO2_{sce}_ppmv.txt"
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Load the base file and new file
base_data = np.loadtxt(base_file, skiprows=1)  # Skip the header
new_data = np.loadtxt(new_file, skiprows=1)    # Skip the header
# Extract time and CO2 values
base_time = base_data[:, 0]
base_co2 = base_data[:, 1]
new_time = new_data[:, 0]
new_co2 = new_data[:, 1]
diff_co2 = new_co2 - base_co2
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(base_time, base_co2, label='Base File')
plt.plot(new_time, new_co2, label='New File')
plt.xlabel('Time')
plt.ylabel('CO2 (ppmv)')
plt.xlim(0, 10000)  # Adjust x-axis limits as needed
# plt.yscale('log')
plt.title('Comparison of CO2 Time Series')
plt.legend(loc='upper left')
plt.grid()

ax2 = plt.gca().twinx()
ax2.plot(new_time, diff_co2, label='Difference', color='red')
ax2.set_ylabel('Difference (ppmv)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.show()
```
