```python
import numpy as np


orbitsF='ForcingData/Samp_orbits_zero_emissions_upd_1myr_AP.res' 
path='../1_CO2_model/SSP/concentration_ppmv/'
sspCO2F_119='CO2_data_SSP1-19_0-1ma.res'
sspCO2F_126='CO2_data_SSP1-26_0-1ma.res'
sspCO2F_245='CO2_data_SSP2-45_0-1ma.res'
sspCO2F_370='CO2_data_SSP3-70_0-1ma.res'
sspCO2F_585='CO2_data_SSP5-85_0-1ma.res'
sspCO2F_10000='CO2_data_10000pgc_0-1ma.res'

obliquity=np.loadtxt(orbitsF,usecols=1,skiprows=1)
esinw=np.loadtxt(orbitsF,usecols=2,skiprows=1)
ecosw=np.loadtxt(orbitsF,usecols=3,skiprows=1)

sspCO2_119=np.loadtxt(path+sspCO2F_119,usecols=1,comments='#',skiprows=1)
sspCO2_126=np.loadtxt(path+sspCO2F_126,usecols=1,comments='#',skiprows=1)
sspCO2_245=np.loadtxt(path+sspCO2F_245,usecols=1,comments='#',skiprows=1)
sspCO2_370=np.loadtxt(path+sspCO2F_370,usecols=1,comments='#',skiprows=1)
sspCO2_585=np.loadtxt(path+sspCO2F_585,usecols=1,comments='#',skiprows=1)
sspCO2_10000=np.loadtxt(path+sspCO2F_10000,usecols=1,comments='#',skiprows=1)

print('[Check]')
print('first 5 values of obliquity:', obliquity[:5])
print('first 5 values of esinw:', esinw[:5])
print('first 5 values of ecosw:', ecosw[:5])
print('first 5 values of sspCO2_126:', sspCO2_126[:5])
print('first 5 values of sspCO2_10000:', sspCO2_10000[:5])
```

```python
# create gsl with value of zero and the same size as other forcings
gsl_zero = np.zeros_like(obliquity)
```

```python
dataSSP119 = np.column_stack((sspCO2_119, obliquity, esinw, ecosw, gsl_zero))
dataSSP126 = np.column_stack((sspCO2_126, obliquity, esinw, ecosw, gsl_zero))
dataSSP245 = np.column_stack((sspCO2_245, obliquity, esinw, ecosw, gsl_zero))
dataSSP370 = np.column_stack((sspCO2_370, obliquity, esinw, ecosw, gsl_zero))
dataSSP585 = np.column_stack((sspCO2_585, obliquity, esinw, ecosw, gsl_zero))
data10000PGC = np.column_stack((sspCO2_10000, obliquity, esinw, ecosw, gsl_zero))
```

```python
# save to file according to the scenarios SSP1-19, SSP1-26, SSP2-45, SSP3-70, SSP4-60, SSP5-34, SSP5-85, 10000PGC
np.savetxt('ForcingData/Samp_orbits_SSP1-19_upd_1myr_AP.res', dataSSP119, header='co2 obliquity esinw ecosw ice', fmt=['%.2f', '%.2f', '%.6f', '%.6f', '%.2f'], comments='')
np.savetxt('ForcingData/Samp_orbits_SSP1-26_upd_1myr_AP.res', dataSSP126, header='co2 obliquity esinw ecosw ice', fmt=['%.2f', '%.2f', '%.6f', '%.6f', '%.2f'], comments='')
np.savetxt('ForcingData/Samp_orbits_SSP2-45_upd_1myr_AP.res', dataSSP245, header='co2 obliquity esinw ecosw ice', fmt=['%.2f', '%.2f', '%.6f', '%.6f', '%.2f'], comments='')
np.savetxt('ForcingData/Samp_orbits_SSP3-70_upd_1myr_AP.res', dataSSP370, header='co2 obliquity esinw ecosw ice', fmt=['%.2f', '%.2f', '%.6f', '%.6f', '%.2f'], comments='')
np.savetxt('ForcingData/Samp_orbits_SSP5-85_upd_1myr_AP.res', dataSSP585, header='co2 obliquity esinw ecosw ice', fmt=['%.2f', '%.2f', '%.6f', '%.6f', '%.2f'], comments='')
np.savetxt('ForcingData/Samp_orbits_10000PGC_upd_1myr_AP.res', data10000PGC, header='co2 obliquity esinw ecosw ice', fmt=['%.2f', '%.2f', '%.6f', '%.6f', '%.2f'], comments='')
```
