这个脚本是把更新后的CO2集合到emul_inputs_SSPxxx.xx.res文件中，确保最后运行的emulator用的是更新后的GSL和CO2

```python
import os

import numpy as np

scenarios=["SSP119","SSP126","SSP245","SSP370","SSP460","SSP585","10000PGC"] 
members=90


# Files read in:
# emul_inputs_SSPxxx.xx.res
# updated_CO2_from_SL_1myr_AP_from500kyrBP.x.res
for mem in range(1,members+1):
# for mem in range(63,65):
    filenameCO2=f"results/updated_CO2_LHS/updated_CO2_from_SL_1myr_AP_from500kyrBP.{mem}.res"
    with open(filenameCO2,'r') as f:
        data_CO2=f.readlines()

    for scenario in scenarios[-1:]:
        filename=f"results/emul_inputs_raw/emul_inputs_{scenario}.{mem}.res"
        with open(filename,'r') as f:
            data_orig=f.readlines()
        scenario_line_index = scenarios.index(scenario)+1
        co2_array = np.array([line.strip().split() for line in data_CO2[2:]], dtype=float)
        CO2 = co2_array[:, scenario_line_index]
        
        # Keep header separate and convert only data rows to float
        header = data_orig[0]
        data = np.array([line.strip().split() for line in data_orig[1:]], dtype=float)
        
        # Update CO2 column (first column, index 0)
        # shape check
        print(f'shape of original data: {data.shape}')
        data[:, 0] = CO2
        
        path_out="results/emul_inputs/"
        if not os.path.exists(path_out):
            os.makedirs(path_out)
        fileout=f"{path_out}emul_inputs_{scenario}.{mem}.updated.res"
        with open(fileout,'w') as f:
            f.write(header)
            for row in data:
                f.write(' '.join(map(str, row)) + '\n')
        print(f"Written updated file: {fileout}")
    
```
