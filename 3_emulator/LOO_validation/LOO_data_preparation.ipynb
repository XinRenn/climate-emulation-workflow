这个脚本主要用于准备ML emulator LOO validation的数据准备
先运行10个看看表现如何。

```python
path = '/Users/bo20541/Library/CloudStorage/OneDrive-UniversityofBristol/TONIC-Oligocene/paleo-emu/examples/training_data/'
modhighice_all_X=path+'training_data_highmodice_temp_formatted.res'
modhighice_all_Y=path+'training_data_highmodice_temp_formatted.nc'
```

```python
import numpy as np
import xarray as xr
X=np.loadtxt(modhighice_all_X)
ds = xr.open_dataset(modhighice_all_Y)
Y = ds['var'].values  # assuming 'temp' is the variable name in the

print(f'X shape: {X.shape}, Y shape: {Y.shape}')
```

```python
# separate ith data of X and save it as validation data, the rest as training data
# separate ith data of Y and save it as validation data, the rest as training data


for i in range(122):
    # separate validation data
    X_val = X[i, :].reshape(1, -1)
    Y_val = Y[i, :, :].reshape(1, Y.shape[1], Y.shape[2])
    # separate training data
    X_train = np.delete(X, i, axis=0)
    Y_train = np.delete(Y, i, axis=0)
    # save validation data
    np.savetxt(f'training_data/emul_in_X_modhigh_{i}.res', X_val)
    # save training data
    np.savetxt(f'training_data/emul_in_X_train_modhigh_{i}.res', X_train)
    # save Y_train as netCDF
    Y_train_ds = xr.Dataset(
        {
            'var': (('sample', 'lat', 'lon'), Y_train)
        },
        coords={
            'sample': np.arange(Y_train.shape[0]),
            'lat': ds['lat'].values,
            'lon': ds['lon'].values
        }
    )
    Y_train_ds.to_netcdf(f"training_data/emul_in_Y_train_modhigh_{i}.nc")
```
