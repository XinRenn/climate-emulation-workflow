```python
import numpy as np
import matplotlib.pyplot as plt

def to_array(data):
    return np.asarray(data, dtype=float)

def as_cell6_log_co2(data):
    arr = to_array(data)
    if np.nanmax(arr) > 30:
        return np.log(arr)
    return arr

# ---- Full-series arrays (same value basis as cell 6) ----
naturaCO2_full = as_cell6_log_co2(naturaCO2)
ssp126CO2_full = as_cell6_log_co2(ssp126CO2)
ssp245CO2_full = as_cell6_log_co2(ssp245CO2)
ssp370CO2_full = as_cell6_log_co2(ssp370CO2)
ssp460CO2_full = as_cell6_log_co2(ssp460CO2)
ssp534CO2_full = as_cell6_log_co2(ssp534CO2)
ssp585CO2_full = as_cell6_log_co2(ssp585CO2)
pgc10000CO2_full = as_cell6_log_co2(pgc10000CO2)

naturagsl_full = to_array(naturagsl)
ssp126gsl_full = to_array(ssp126gsl)
ssp245gsl_full = to_array(ssp245gsl)
ssp370gsl_full = to_array(ssp370gsl)
ssp460gsl_full = to_array(ssp460gsl)
ssp534gsl_full = to_array(ssp534gsl)
ssp585gsl_full = to_array(ssp585gsl)
pgc10000gsl_full = to_array(pgc10000gsl)

# ---- Irregular X display with different resolution by interval ----
anchor_years = np.concatenate([
    np.arange(0, 11, 2),
    np.arange(20, 101, 20),
    np.arange(200, 1001, 50)
]).astype(float)
anchor_pos = np.arange(len(anchor_years), dtype=float)

full_years = np.arange(naturaCO2_full.shape[1], dtype=float)
x_display = np.interp(full_years, anchor_years, anchor_pos)

# Visual boundaries for interval separation at x=10 and x=100 (year space)
x10 = np.interp(10, anchor_years, anchor_pos)
x100 = np.interp(100, anchor_years, anchor_pos)

plt.figure(figsize=(12, 7))

# CO2 plot
ax1 = plt.subplot(2, 1, 1)
ax1.axvspan(x_display[0], x10, color='lightgrey', alpha=0.24, zorder=0)
ax1.axvspan(x10, x100, color='lightyellow', alpha=0.20, zorder=0)
ax1.axvspan(x100, x_display[-1], color='lightcyan', alpha=0.20, zorder=0)
for i in range(90):
    plt.plot(x_display, naturaCO2_full[i], color='grey', alpha=0.01)
    plt.plot(x_display, ssp126CO2_full[i], color='lightblue', alpha=0.01)
    plt.plot(x_display, ssp245CO2_full[i], color='lightgreen', alpha=0.01)
    plt.plot(x_display, ssp370CO2_full[i], color='khaki', alpha=0.01)
    # plt.plot(x_display, ssp460CO2_full[i], color='peachpuff', alpha=0.01)
    # plt.plot(x_display, ssp534CO2_full[i], color='lightpink', alpha=0.01)
    plt.plot(x_display, ssp585CO2_full[i], color='lightcoral', alpha=0.01)
    plt.plot(x_display, pgc10000CO2_full[i], color='plum', alpha=0.01)
plt.plot(x_display, naturaCO2_full[66], color='black', label='Natural', linewidth=1)
plt.plot(x_display, ssp126CO2_full[66], color='blue', label='SSP126', linewidth=1)
plt.plot(x_display, ssp245CO2_full[66], color='green', label='SSP245', linewidth=1)
plt.plot(x_display, ssp370CO2_full[66], color='orange', label='SSP370', linewidth=1)
# plt.plot(x_display, ssp460CO2_full[66], color='yellow', label='SSP460', linewidth=1)
# plt.plot(x_display, ssp534CO2_full[66], color='pink', label='SSP534', linewidth=1)
plt.plot(x_display, ssp585CO2_full[66], color='red', label='SSP585', linewidth=1)
plt.plot(x_display, pgc10000CO2_full[66], color='darkmagenta', label='10000pgc', linewidth=1)
ax1.axvline(x10, color='white', linestyle='-', linewidth=4)
ax1.axvline(x100, color='white', linestyle='-', linewidth=4)

# Show y-axis labels in actual ppmv while plotting log(CO2) values
ppm_ticks = np.array([180, 280, 400, 560, 840, 1120, 2240, 4480], dtype=float)
ax1.set_yticks(np.log(ppm_ticks))
ax1.set_yticklabels([f"{int(v)}" for v in ppm_ticks])

plt.ylabel('CO2 concentration (ppmv)')
plt.title('Projected CO2 Concentrations under Different SSP Scenarios')
plt.legend()
plt.grid()

# Tick labels at anchor years
plt.xticks(anchor_pos, anchor_years.astype(int), rotation=45)

# GSL plot
ax2 = plt.subplot(2, 1, 2)
ax2.axvspan(x_display[0], x10, color='lightgrey', alpha=0.24, zorder=0)
ax2.axvspan(x10, x100, color='lightyellow', alpha=0.20, zorder=0)
ax2.axvspan(x100, x_display[-1], color='lightcyan', alpha=0.20, zorder=0)
for i in range(90):
    plt.plot(x_display, naturagsl_full[i], color='lightgrey', alpha=0.02)
    plt.plot(x_display, ssp126gsl_full[i], color='lightblue', alpha=0.02)
    plt.plot(x_display, ssp245gsl_full[i], color='lightgreen', alpha=0.02)
    plt.plot(x_display, ssp370gsl_full[i], color='khaki', alpha=0.02)
    # plt.plot(x_display, ssp460gsl_full[i], color='peachpuff', alpha=0.02)
    # plt.plot(x_display, ssp534gsl_full[i], color='lightpink', alpha=0.02)
    plt.plot(x_display, ssp585gsl_full[i], color='lightcoral', alpha=0.02)
    plt.plot(x_display, pgc10000gsl_full[i], color='plum', alpha=0.02)
plt.plot(x_display, naturagsl_full[66], color='black', label='Natural', linewidth=1)
plt.plot(x_display, ssp126gsl_full[66], color='blue', label='SSP126', linewidth=1)
plt.plot(x_display, ssp245gsl_full[66], color='green', label='SSP245', linewidth=1)
plt.plot(x_display, ssp370gsl_full[66], color='orange', label='SSP370', linewidth=1)
# plt.plot(x_display, ssp460gsl_full[66], color='yellow', label='SSP460', linewidth=1)
# plt.plot(x_display, ssp534gsl_full[66], color='pink', label='SSP534', linewidth=1)
plt.plot(x_display, ssp585gsl_full[66], color='red', label='SSP585', linewidth=1)
plt.plot(x_display, pgc10000gsl_full[66], color='darkmagenta', label='10000pgc', linewidth=1)
plt.plot(x_display, np.full_like(x_display, -28), color='black', linewidth=3, linestyle='dashed', alpha=0.7)
ax2.axvline(x10, color='white', linestyle='-', linewidth=4)
ax2.axvline(x100, color='white', linestyle='-', linewidth=4)
plt.xlabel('Time (kyr)')
plt.ylabel('Global Sea Level (m)')
plt.title('Projected GSL under Different SSP Scenarios')
plt.grid()

# Tick labels at anchor years
plt.xticks(anchor_pos, anchor_years.astype(int), rotation=45)

plt.tight_layout()
plt.savefig('CO2_GSL_SSP.png', dpi=300)
plt.show()

# Check consistency at year 1000 against full series
print('Check member 67 @ year 1000 (CO2, log):', naturaCO2_full[66, 1000])
print('Check member 67 @ year 1000 (GSL):', naturagsl_full[66, 1000])
```
